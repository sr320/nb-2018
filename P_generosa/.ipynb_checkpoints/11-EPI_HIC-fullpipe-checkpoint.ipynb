# Mapping

```python
%%bash
find EPI*_*R1*.fastq.gz \
| xargs basename -s _L005_R1_001.fastq.gz | xargs -I{} /Applications/bioinfo/Bismark_v0.19.0/bismark \
--path_to_bowtie /Applications/bioinfo/bowtie2-2.3.4.1-macos-x86_64 \
--genome /Volumes/Orange_block/data/18-05-02/Results/geoduck_roberts2018-04-21/ \
--score_min L,0,-1.2 \
-u 10000 \
-p 1 \
--non_directional \
-1 {}_L005_R1_001.fastq.gz \
-2 {}_L005_R2_001.fastq.gz \
2> bismark.err
```
